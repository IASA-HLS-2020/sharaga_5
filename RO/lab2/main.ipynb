```python
# from skfuzzy import cmeans
from matplotlib import image
from pathlib import Path
from sklearn.cluster import KMeans
import  matplotlib.pyplot as plt


processed_images = [image.imread(file) / 255 for file in Path.cwd().glob('data/*.jpg')]

for pr_img_id, pr_img in enumerate(processed_images):

    x, y ,z = pr_img.shape

    clf = KMeans(n_clusters=7)
    clf.fit(pr_img.reshape(x * y, z))

    plt.imshow(clf.cluster_centers_[clf.labels_].reshape(x,y,z))
    img = clf.cluster_centers_[clf.labels_].reshape(x,y,z)
    plt.figure()
    plt.imshow(img)
#
```
